```python
import os, argparse
import ijson
import re
import string
import pickle
```

```python
def read_list_from_text(the_filename):
    with open(the_filename, 'rb') as f:
        my_list = pickle.load(f)
    print('Reading list of unique words from file ...........')
    return my_list
```

```python
mylist=(read_list_from_text('Unique_Words_Color_Test.txt') )
    
```

```python
print(mylist)
```

```python
def ConvertTupleToList(Tuple):
    tupleToList = []
    for item in Tuple:
        tupleToList.append(item[0])

    return tupleToList
```

```python
newlist=ConvertTupleToList(mylist)
```
